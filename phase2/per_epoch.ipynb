```python
import pandas as pd
import numpy as np

# =====================================================
# CONSTANTS
# =====================================================
EPOCHS = 4
IMAGES_PER_EPOCH = 45000
```

```python
# =====================================================
# RAW ENERGY DATA (kWh)
# =====================================================
data = {
    "EfficientNet": [0.021644, 0.021875, 0.021842, 0.021869, 0.021886],
    "MobileNet": [0.020362, 0.020701, 0.020826, 0.020965, 0.021061],
    "ShuffleNet": [0.028161, 0.027674, 0.027780, 0.027839, 0.027983],
    "MESNet (Ensemble)": [0.014579, 0.014518, 0.014441, 0.014666, 0.014898],
}
```

```python
# =====================================================
# HELPER FUNCTION (for 10⁻ⁿ format)
# =====================================================
def format_10power(x):
    if x == 0:
        return "0"
    exponent = int(np.floor(np.log10(abs(x))))
    base = x / (10 ** exponent)
    return f"{base:.3f} ×10^{exponent}"
```

```python
# =====================================================
# CALCULATIONS
# =====================================================
rows = []

for model, energy_values in data.items():
    # Mean and Standard Deviation
    mean_energy = np.mean(energy_values)
    sd_energy = np.std(energy_values, ddof=1)

    # Per epoch (4 epochs total)
    energy_per_epoch = mean_energy / EPOCHS

    # Per 1000 images (each epoch = 45,000 images)
    energy_per_1000 = energy_per_epoch / (IMAGES_PER_EPOCH / 1000)

    # Per inference (1 image)
    energy_per_inference = energy_per_epoch / IMAGES_PER_EPOCH

    # Store results with formatted powers of 10
    rows.append([
        model,
        f"{mean_energy:.6f} ± {sd_energy:.6f}",
        format_10power(energy_per_epoch),
        format_10power(energy_per_1000),
        format_10power(energy_per_inference)
    ])
```

```python

# =====================================================
# CREATE TABLE
# =====================================================
columns = [
    "Model",
    "Energy (mean ± sd, kWh)",
    "Energy / Epoch (kWh)",
    "Energy / 1000 Images (kWh)",
    "Energy / Inference (kWh)",
]

df = pd.DataFrame(rows, columns=columns)
```

```python
# =====================================================
# DISPLAY CLEAR TABLE
# =====================================================
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 200)
print("\n======================== ENERGY SUMMARY (10⁻ⁿ Format) ========================\n")
print(df.to_string(index=False))
print("\n=============================================================================\n")
```
