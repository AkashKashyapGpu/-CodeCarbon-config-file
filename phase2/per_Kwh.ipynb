```python
import pandas as pd
import numpy as np

# -------------------------------------------------------------
# 1️⃣ Input Mean ± SD Data
# -------------------------------------------------------------
data = {
    "Model": [
        "EfficientNet V2-S",
        "ShuffleNet V2",
        "MobileNet V3",
        "Ensemble MESNET"
    ],
    "Accuracy": [92.08, 91.74, 92.56, 96.98],
    "Precision": [92.12, 91.76, 92.66, 97.01],
    "Recall": [92.08, 91.73, 92.56, 96.98],
    "F1_Score": [92.07, 91.72, 92.57, 96.98],
    "Energy_kWh": [0.021823, 0.027887, 0.020783, 0.014620],
    "CO2_kg": [0.007621, 0.009739, 0.009784, 0.006883],
    "GFLOPs": [0.184, 0.044, 0.006, 0.146],
}

df = pd.DataFrame(data)

# -------------------------------------------------------------
# 2️⃣ Derived Metrics — but without EEI and EATS
# -------------------------------------------------------------
df["Accuracy_per_kWh"]  = df["Accuracy"]  / df["Energy_kWh"]
df["Precision_per_kWh"] = df["Precision"] / df["Energy_kWh"]
df["Recall_per_kWh"]    = df["Recall"]    / df["Energy_kWh"]
df["F1_per_kWh"]        = df["F1_Score"]  / df["Energy_kWh"]

# -------------------------------------------------------------
# 3️⃣ Display Results — Full precision (no rounding)
# -------------------------------------------------------------
pd.set_option("display.float_format", "{:.6f}".format)

print("=== Derived Performance and Energy Efficiency Metrics (Mean Values) ===\n")
display(df[[
    "Model", "Accuracy", "Precision", "Recall", "F1_Score",
    "Energy_kWh", "CO2_kg", "GFLOPs",
    "Accuracy_per_kWh", "Precision_per_kWh", "Recall_per_kWh", "F1_per_kWh"
]])
```
